```python
expt = 'DAv7_M36_SMAP_type_10_fp'

path = f'/discover/nobackup/amfox/Experiments/{expt}/{expt}/output/SMAP_EASEv2_M36_GLOBAL/ana/ens_avg/Y2015/M04'
# path = f'/discover/nobackup/amfox/Experiments/{expt}/{expt}/scratch'

file_name_start = f'{expt}.ens_avg.ldas_ObsFcstAna.20150401_03'
printflag = False

date_time2, obs_species2, obs_tilenum2, obs_lon2, obs_lat2, obs_obs2, obs_obsvar2, obs_fcst2, obs_fcstvar2, obs_ana2, obs_anavar2 = read_obsfcstana(path, file_name_start, printflag)

numobs2 = len(obs_obs2[obs_species2 < 5])
obarray2 = np.empty([numobs2, 3])
obarray2[:, 1] = obs_lon2[obs_species2 < 5]
obarray2[:, 2] = obs_lat2[obs_species2 < 5]
obarray2[:, 0] = obs_obs2[obs_species2 < 5]

plot_global_tight(obarray2,False,f'Observations: {expt}','(Tb)')
```

```python
obs_ana[obs_ana < -9998] = np.nan
obs_ana2[obs_ana2 < -9998] = np.nan

print('nobs = ',numobs, numobs2)
print('Mean obs =', np.mean(obs_obs[obs_species < 5]), np.mean(obs_obs2[obs_species2 < 5]))

print('Mean Fcst =', np.mean(obs_fcst[obs_species < 5]), np.mean(obs_fcst2[obs_species2 < 5]))
print('Max Fcst =', np.max(obs_fcst[obs_species < 5]), np.max(obs_fcst2[obs_species2 < 5]))
print('Min Fcst =', np.min(obs_fcst[obs_species < 5]), np.min(obs_fcst2[obs_species2 < 5]))

print('Mean Ana =', np.mean(obs_ana[obs_species < 5]), np.nanmean(obs_ana2[obs_species2 < 5]))
print('Max Ana =', np.max(obs_ana[obs_species < 5]), np.nanmax(obs_ana2[obs_species2 < 5]))
print('Min Ana =', np.min(obs_ana[obs_species < 5]), np.nanmin(obs_ana2[obs_species2 < 5]))


print('Mean OmF =', np.mean(obs_obs[obs_species < 5] - obs_fcst[obs_species < 5]), np.mean(obs_obs2[obs_species2 < 5] - obs_fcst2[obs_species2 < 5]))
print('Max OmF =', np.max(obs_obs[obs_species < 5] - obs_fcst[obs_species < 5]), np.max(obs_obs2[obs_species2 < 5] - obs_fcst2[obs_species2 < 5]))
print('Min OmF =', np.min(obs_obs[obs_species < 5] - obs_fcst[obs_species < 5]), np.min(obs_obs2[obs_species2 < 5] - obs_fcst2[obs_species2 < 5]))

print('Mean increment = ', np.mean(obs_ana[obs_species < 5] - obs_fcst[obs_species < 5]), np.nanmean(obs_ana2[obs_species2 < 5] - obs_fcst2[obs_species2 < 5]))
print('Max increment = ', np.max(obs_ana[obs_species < 5] - obs_fcst[obs_species < 5]), np.nanmax(obs_ana2[obs_species2 < 5] - obs_fcst2[obs_species2 < 5]))
print('Min increment = ', np.min(obs_ana[obs_species < 5] - obs_fcst[obs_species < 5]), np.nanmin(obs_ana2[obs_species2 < 5] - obs_fcst2[obs_species2 < 5]))
```
