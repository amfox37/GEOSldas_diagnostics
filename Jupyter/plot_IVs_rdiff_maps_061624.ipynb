```python
import scipy.io as sio
import numpy as np

from mapper_functions import plot_global_tight
```

```python
# Open .mat files
data = sio.loadmat('../test_data/fp_scaled/Rdiff_DAv7_M36_ASCAT_type_13_comb_fp_scaled_minus_OLv7_M36_MULTI_type_13_comb_fp_scaled.mat')

# Get the data from the .mat file which contains 3 variables, Rdiff_vector, lons, lats
Rdiff_vector = np.squeeze(data['Rdiff_vector'])
lons = np.squeeze(data['lons'])
lats = np.squeeze(data['lats'])

ob_array = np.empty([len(Rdiff_vector),3])
ob_array[:,1] = lons
ob_array[:,2] = lats
ob_array[:,0] = Rdiff_vector

# Plot the data
plot_global_tight(ob_array,True,'ASCAT only minus OL: Surface Soil Moisture Skill (Î” anomaly R)','(-)', -0.5, 0.5)

print("Mean:" + str(np.nanmean(Rdiff_vector)))
Rdiff_vector_ASCT = Rdiff_vector
```

```python
# Open .mat files
# data = sio.loadmat('../test_data/Rdiff_DAv7_M36_ASCAT_type_13_comb_fp_minus_OLv7_M36_FPprcp.mat')
data = sio.loadmat('../test_data/fp_scaled/Rdiff_DAv7_M36_SMAP_type_13_comb_fp_scaled_minus_OLv7_M36_MULTI_type_13_comb_fp_scaled.mat')

# Get the data from the .mat file which contains 3 variables, Rdiff_vector, lons, lats
Rdiff_vector = np.squeeze(data['Rdiff_vector'])
lons = np.squeeze(data['lons'])
lats = np.squeeze(data['lats'])

ob_array = np.empty([len(Rdiff_vector),3])
ob_array[:,1] = lons
ob_array[:,2] = lats
ob_array[:,0] = Rdiff_vector

# Plot the data
plot_global_tight(ob_array,True,'SMAP only minus OL: Surface Soil Moisture Skill (Î” anomaly R)','(-)', -0.5, 0.5)

print("Mean:" + str(np.nanmean(Rdiff_vector)))
Rdiff_vector_SMAP = Rdiff_vector
```

```python
# Plot the PDFs of the different Rdiff vectors
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

plt.figure(figsize=(16, 8))  
sns.kdeplot(Rdiff_vector_ASCT, label='ASCT')
sns.kdeplot(Rdiff_vector_SMAP, label='SMAP')
plt.xlabel('Î” anomaly R')
plt.ylabel('Density')
plt.title('PDF of Î” anomaly R', fontsize=24)
plt.legend()
plt.savefig('Rdiff_PDFs.png')
plt.show()
```
